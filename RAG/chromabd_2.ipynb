```python
import os
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer

# Initialize the embedding model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Initialize the ChromaDB client
client = chromadb.PersistentClient(path="./.chroma_db")
collection = client.get_or_create_collection(name="my_text_collection")

# Function: Chunk text
def chunk_text(text, chunk_size=300, overlap=50):
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i:i + chunk_size])
        chunks.append(chunk)
    return chunks

# Function: Get new files in the folder
def get_new_files(folder_path, collection):
    # Get all file names in the folder
    all_files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]

    # Get file names already in the database
    existing_files = [meta['source'] for meta in collection.get()['metadatas']]

    # Find new files
    new_files = [f for f in all_files if f not in existing_files]
    return new_files

# Function: Process and add new files to the database
def process_and_add_new_files(folder_path, collection):
    new_files = get_new_files(folder_path, collection)
    if not new_files:
        return False  # No new files to process

    for new_file in new_files:
        with open(os.path.join(folder_path, new_file), 'r') as file:
            content = file.read()

        # Chunk and embed the file content
        chunks = chunk_text(content)
        embeddings = embedding_model.encode(chunks).tolist()

        # Create unique IDs and metadata
        ids = [f"{new_file}_chunk_{i}" for i in range(len(chunks))]
        metadata = [{"source": new_file} for _ in range(len(chunks))]

        # Add to the database
        collection.upsert(
            documents=chunks,
            embeddings=embeddings,
            metadatas=metadata,
            ids=ids
        )
        print(f"Added {new_file} to the database.")

    return True  # New files were processed

# Function: Query the vector database
def query_database(collection, query_text, n_results=3):
    query_embedding = embedding_model.encode([query_text]).tolist()
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=n_results
    )
    print("\nQuery Results:")
    for doc, meta in zip(results['documents'][0], results['metadatas'][0]):
        print(f"Document: {doc[:50]}... (truncated)\nMetadata: {meta}\n")

# Function: Inspect database contents
def inspect_database(collection):
    all_data = collection.get()
    print("\nDatabase Contents:")
    for doc, meta in zip(all_data['documents'], all_data['metadatas']):
        print(f"Document: {doc[:50]}... (truncated)\nMetadata: {meta}\n")

# Folder path containing text files
folder_path = "./text_files"

# Ensure the folder exists
os.makedirs(folder_path, exist_ok=True)

# Main Workflow
def main_workflow(folder_path, collection):
    # Check and process new files
    new_files_added = process_and_add_new_files(folder_path, collection)

    # If no new files, notify the user
    if not new_files_added:
        print("No new files detected. Using the current database.")

    # Inspect the current state of the database (Optional)
    inspect_database(collection)

    # Query the database
    query_text = input("\nEnter your query: ")
    query_database(collection, query_text=query_text, n_results=3)

# Run the workflow
main_workflow(folder_path, collection)
```
